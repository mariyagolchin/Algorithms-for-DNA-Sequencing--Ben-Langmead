```
import os
def readGenome(file_name):
    genome=''
    with open(file_name, 'r') as f:
       for line in f:
          # print(line)
          if not line[0] == '>' :
             print(line)
             genome += line.rstrip()
    return genome

path='/content/drive/My Drive/Data/lambda_virus.fa'
genome=readGenome(path)
print(genome)
print(len(genome))
```
